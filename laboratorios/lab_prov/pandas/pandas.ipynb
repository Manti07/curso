```python
avocado_df.groupby('region')['Total Bags'].sum().plot(kind='bar')
```

```python
avocado_df.groupby(['region','year'])['Total Bags'].sum().plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

avocado_df

avocados_region_year = avocado_df.groupby(['region','year'])['Total Bags'].sum()
avocados_region_year_unstacked = avocados_region_year.unstack()

important_sales = avocados_region_year_unstacked[avocados_region_year_unstacked.sum(axis=1) > 1*10**7]

important_sales.plot(kind='bar', color=['skyblue', 'salmon', 'orange' , 'red'], figsize=(10,6), xlabel='Region', grid=True)
```

```python
avocado_df

avocados_region_mean = avocado_df.groupby('region')['AveragePrice'].mean()
avocados_region_mean.plot(kind='bar', grid= True)
```

```python
avocados_relative_mean = (avocados_region_mean - avocados_region_mean.mean()/avocados_region_mean)
avocados_relative_mean.plot(kind = 'bar')
```
